# Loading and exploring the dataset

```python
from sklearn.datasets import fetch_openml
import numpy as np

# Fetch MNIST dataset
mnist = fetch_openml('mnist_784', version=1, as_frame=False)  # Optional but smart
X, y = mnist["data"], mnist["target"]

# Convert to NumPy arrays (if not already NumPy due to `as_frame=False`)
X = np.array(X)
y = np.array(y, dtype=int)  # Make sure y is integer (not string)

# Now everything is NumPy and happy ðŸ˜Œ
print(X.shape)  # (70000, 784)
print(y[0])     # e.g., 5
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
print(X.shape)

for i in range(0,10):
    print(y[i])
    someDigitArray = X[i]
    someDigitImage = someDigitArray.reshape(28,28)
    plt.imshow(someDigitImage,cmap='binary')
    plt.axis("off")
    plt.show()
```
